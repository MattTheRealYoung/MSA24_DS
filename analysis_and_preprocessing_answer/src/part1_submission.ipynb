Convert non-numerical and categorical columns to numerical ones. By looking at the columns' data type, we applied the change to `object` data type except `Order Date` and `Ship Date` as they can be converted to timestamp. Meanwhile, declared a new variable to store the changed columns for later using.

```python
# Convert non-numerical columns into numerical using label encoding
label_encoder = LabelEncoder()

#Record the processed columns in converted data frame
converted_df = pd.DataFrame()
for col in df.select_dtypes(include=['object']).columns:
    # Skip Order Date and Ship Date which can be converted to timestamp
    if col.startswith('Order Date') or col.startswith('Ship Date'):
        continue
    df[col] = label_encoder.fit_transform(df[col])
    converted_df[col] = df[col].copy()
```

### 2. Clean data


At this step, we will check is there any missing values in the datasets, remove or input a reasonable value into particular attribute. Then remove the outliers which helps our model learn better.

```python
# Check the data set if there is any missing value
missing_data = df.isnull().sum().to_frame()
print(missing_data)
```

As there is no missing values, we can skip drop the missing values or insert a value for missing columns. Next, Let's look at the outliers

```python
# Identify outliers using IQR, as Order Date and Ship Date are string, skip first 4 columns
df_quantile_columns = df.iloc[:, 4:]
q1 = df_quantile_columns.quantile(0.25)
q3 = df_quantile_columns.quantile(0.75)
iqr = q3 - q1
outliers = (df_quantile_columns < (q1 - 1.5 * iqr)) | (df_quantile_columns > (q3 + 1.5 * iqr))
print(outliers.sum())

#Removing Sales, Quantity, Profit outliers
q1_sales = df['Sales'].quantile(0.25)
q3_sales = df['Sales'].quantile(0.75)
iqr_sales = q3_sales - q1_sales
lower_sales = q1_sales - 1.5 * iqr_sales
upper_sales = q3_sales + 1.5 * iqr_sales
cleaned_df = df[(df['Sales'] >= lower_sales) & (df['Sales'] <= upper_sales)]

q1_quantity = cleaned_df['Quantity'].quantile(0.25)
q3_quantity = cleaned_df['Quantity'].quantile(0.75)
iqr_quantity = q3_quantity - q1_quantity
lower_quantity = q1_quantity - 1.5 * iqr_quantity
upper_quantity = q3_quantity + 1.5 * iqr_quantity
cleaned_df = cleaned_df[(cleaned_df['Quantity'] >= lower_quantity) & (cleaned_df['Quantity'] <= upper_quantity)]

q1_profit = cleaned_df['Profit'].quantile(0.25)
q3_profit = cleaned_df['Profit'].quantile(0.75)
iqr_profit = q3_profit - q1_profit
lower_profit = q1_profit - 1.5 * iqr_profit
upper_profit = q3_profit + 1.5 * iqr_profit
cleaned_df = cleaned_df[(cleaned_df['Profit'] >= lower_profit) & (cleaned_df['Profit'] <= upper_profit)]
df = cleaned_df
```

Some features are unrelated to our context and will influence our prediction, so drop non-relative features

```python
# Drop the unrelated features
df = df.drop(columns=['Row ID', 'Order ID', 'Ship Date', 'Ship Mode', 'Customer ID', 'Customer Name', 'Segment', 'Country', 'City', 'State', 'Postal Code', 'Region', 'Product ID', 'Category', 'Sub-Category', 'Product Name'])
df
```

```python
# Format Order Date to datetime, aggregate Sales data to weekly and monthly
df.columns = df.columns.str.strip()
print(df.columns)
if 'Order Date' in df.columns:
    df['Order Date'] = pd.to_datetime(df['Order Date'], format='%m/%d/%y')
else:
    raise KeyError('The expected Column Order Date is missing')
df.set_index('Order Date', inplace=True)
weekly_sales = df['Sales'].resample('W').sum()
monthly_sales = df['Sales'].resample('ME').sum()
quarterly_sales = df['Sales'].resample('QE').sum()
yearly_sales = df['Sales'].resample('YE').sum()
```

### 4. Identify correlated variables
Using heatmap, we can find the correlation of each variable relates to the others. And, we can identify and highlight the apparent relationships between variables. 

```python
#Compute the correlation matrix
corr_matrix = df.corr()

#Plot the heatmap
plt.figure(figsize=(14, 10))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix")
plt.show()
```

From the matrix, we can see:
- Sales has the strongest relationship with quantity,as quantity decided how many we can sell and will influence sales
- Sales has some connection with profit, as profit come from sales
- Profit has strong negative relationship with Discount, as the more discount of the order the less profit the order can gain
